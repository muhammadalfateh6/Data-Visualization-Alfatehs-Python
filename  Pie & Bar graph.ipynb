```python
import matplotlib.pyplot as plt
import numpy as np

values = [5411 , 1109  ]
labels = ['Benign', 'Malignant']
colorss=[ 'tab:blue', 'tab:red']
x = np.arange(len(labels))
plt.bar(x, values, color=colorss, width=0.5)
plt.xticks(x, labels)
plt.ylabel('Number Of Images')
plt.title('Class Distribution In HAM10000 Dataset')
```

```python
import matplotlib.pyplot as plt

values = [12.5, 48, 13.5, 13, 13 ]
labels = ['ISIC2016', 'PH2','ISIC2018','ISIC2020','HAM10000']
colorss=[ 'tab:blue','tab:orange','tab:green', 'tab:red', 'tab:purple']
plt.subplots(figsize=(4,4 ))
plt.pie(values, labels=labels, textprops={'weight':'bold'}, startangle=90, autopct='%1.1f%%', colors=colorss)

plt.show()
```
